```python
from aioredis import create_redis_pool
import aioredis_cluster
from common import ENV
from urllib.parse import urlparse
from pyprnt import prnt
```

```python

"""
if 'REDIS_USERNAME' in ENV and 'REDIS_PASSWORD' in ENV:
    red = urlparse(ENV.REDIS_URL)
    redis_url = f"{red.scheme}://{ENV.REDIS_USERNAME}:{ENV.REDIS_PASSWORD}@{red.netloc}"
else:
    redis_url = ENV.REDIS_URL
"""

async def main():
    redis_url = ENV.REDIS_URL
    print('Connecting redis: ', redis_url)

    '''cluster_pool = await aioredis_cluster.create_redis_cluster([
        redis_url,
    ])'''
    redis_pool = await create_redis_pool(redis_url,
                                  minsize=5,
                                  maxsize=100)
    ch, *_ = await redis_pool.subscribe("globe:surveys")
    async for msg in ch.iter(encoding="utf-8"):
        prnt(msg)


await main()
```
